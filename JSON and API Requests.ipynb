# JSON and API Requests

## Setup

```python
import requests
import json
```

## JSON to Python

```python
x =  '{ "name":"John", "age":30, "city":"New York"}'
y = json.loads(x)   # parse x
print(y)            # the result is a Python dictionary
print(y["age"])     # access dictionary values
```

## Python to JSON

```python
x = {
  "name": "John",
  "age": 30,
  "city": "New York"
}
y = json.dumps(x)   # convert into JSON
print(y)            # the result is a JSON string
```

## JSON Request Example

```python
url = 'http://pokeapi.co/api/v2/pokemon/?limit=5000'
request = requests.get(url)
data = json.loads(request.content)

print('Number of Pokemon:',len(data['results']))
print('Pokemon names:')
for pokemon in data['results']:
    print(pokemon['name'])
```

# space X request

```python
url = 'http://api.spacexdata.com/v3/launches/latest?pretty=true'
request = requests.get(url)
data = json.loads(request.content)
rocket_data = data['rocket']
print(rocket_data['second_stage'])
```

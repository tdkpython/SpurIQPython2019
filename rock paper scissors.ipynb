# rock paper scissors

```python
import random 
moves = ["r", "p", "s"]
player_wins = ["pr", "sp", "rs"]
your_score = 0
my_score = 0
while True:
    player_move = input("your move: ")
    computer_move = random.choice(moves)
    
    if player_move == "q":
        break
    print("you: ", player_move)
    print("me: ", computer_move)

    if player_move == computer_move:
        print("tie")
    elif player_move + computer_move in player_wins:
        print("too bad you win")
        your_score  += 1
    else:
        print("haha you lose")
        my_score += 1
         
    if your_score == 20:
        print ("you win")
        break
    if my_score == 20:
        print ("I win")
        break
    print(your_score)
    print(my_score)
```
